```python
import numpy as np
import pandas as pd
import random
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.metrics import mutual_info_score
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression
```

```python
BF_df_CTU = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/BF_df_CTU.feather")
BF_df_OU = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/BF_df_OU.feather")
BF_df_V = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/BF_df_V.feather")
DT_df_CC = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/DT_df_CC.feather")
DT_df_JI = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/DT_df_JI.feather")
IADQ_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/IADQ_df.feather")
IESR_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/IESR_df.feather")
NAQ_R_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/NAQ_R_df.feather")
PCL5_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/PCL5_df.feather")
PHQ9_GAD7_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/PHQ9_GAD7_df.feather")
PID5_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/PID5_df.feather")
PRFQ_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/PRFQ_df.feather")
PRMQ_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/PRMQ_df.feather")
R_NEO_PI = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/R_NEO_PI.feather")
RAW_DDDT = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/RAW_DDDT.feather")
sPID_5_df = pd.read_feather("C:/Users/Asus/Desktop/cognitive_datasets/data/sPID-5_df.feather")
```

```python
datasets = [BF_df_CTU, BF_df_OU,BF_df_V,DT_df_CC,DT_df_JI,IADQ_df,IESR_df,NAQ_R_df,PCL5_df,
            PHQ9_GAD7_df,PID5_df,PRFQ_df,PRMQ_df,R_NEO_PI,RAW_DDDT,sPID_5_df]
results = {}
Xs = []
ys = []
for i in range(len(datasets)):
        
        X = datasets[i].iloc[:,:-1]
        y = datasets[i].iloc[:,-1]
        scaler = StandardScaler()
        scaler.fit(X)
        X = scaler.transform(X)
        Xs.append(X)
        ys.append(y)
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
        
        model1 = LogisticRegression(random_state=1)
        model1.fit(X_train, y_train)
        pred1 = model1.predict(X_test)
        scorelogreg = accuracy_score(pred1,y_test)

        model2 = RandomForestClassifier(random_state=1)
        model2.fit(X_train, y_train)
        pred2 = model2.predict(X_test)
        scorerandfor = accuracy_score(pred2, y_test)

        model3 = LinearSVC(random_state=1,max_iter = 10000)
        model3.fit(X_train, y_train)
        pred3 = model3.predict(X_test)
        scoresvm = accuracy_score(pred3,y_test)

        results[i] = [scorelogreg,scorerandfor,scoresvm]


results
```

```python
mis = {}
for d in range(len(datasets)):
    result = np.array(mutual_info_classif(Xs[d], ys[d]))
    for i in range(10):
        result = np.mean(np.array([result,mutual_info_classif(Xs[d], ys[d])]),axis = 0)
    maxresult = max(result)
    result /= maxresult
    mis[d] = result
    
mis
```

```python
new_results = {}
for d in range(len(datasets)):
    mi = mis[d]
    indices = mi >= 0.75
    X = Xs[d][:,indices]
    y = ys[d]
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
        
    model1 = LogisticRegression(random_state=1)
    model1.fit(X_train, y_train)
    pred1 = model1.predict(X_test)
    scorelogreg = accuracy_score(pred1,y_test)

    model2 = RandomForestClassifier(random_state=1)
    model2.fit(X_train, y_train)
    pred2 = model2.predict(X_test)
    scorerandfor = accuracy_score(pred2, y_test)

    model3 = LinearSVC(random_state=1,max_iter = 10000)
    model3.fit(X_train, y_train)
    pred3 = model3.predict(X_test)
    scoresvm = accuracy_score(pred3,y_test)
    
    new_results[d] = [scorelogreg,scorerandfor,scoresvm]

new_results
```

```python
final = {}
for i in range(len(new_results)):
        final[i] = list(zip(results[i],new_results[i]))
final
```
